```python
import json
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

ROOT = "../../data/deskew_autocrop_normalize"  # Chưa fix folder nên để tạm cái ROOT này cho phù hợp với local thôi nha, từ từ hẵng merge code tại file JSON để test là tự tạo thôi.
fileraw = 'fileraw_test.json'
raw = os.path.join(ROOT, fileraw)

with open(raw, 'r', encoding='utf-8') as f:
    data_sample = json.load(f)

print(data_sample) 
```

```python
def convert_to_bounding_box(raw_bbox):
    x_min, y_min, x_max, y_max = raw_bbox
    
    x = x_min
    y = y_min
    w = x_max - x_min
    h = y_max - y_min
    
    return {"x": x, "y": y, "w": w, "h": h}

# Chạy thử với dữ liệu mẫu vừa đọc
for item in data_sample:
    if "raw_bbox" in item:
        bbox_standard = convert_to_bounding_box(item["raw_bbox"])
        print(f"Text: {item['text']} -> BBox: {bbox_standard}")
```
